# Part B: Time Convolution

1)

```python
def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)

n=np.arange(0,1001,1)
x2=np.sin((2/3)*np.pi*n)*(u(0,n)-u(3,n))
O1=np.linspace(-np.pi,np.pi,1001)
W_O= np.exp(-1j)**np.dot((np.arange(x2.size).reshape(x2.size,1)) , (O1.reshape(1,1001)))
X2=np.dot(x2,W_O)

absX=[abs(Xs) for Xs in X2]
plt.plot(absX)
plt.title('DTFT of x[n]')
plt.xlabel('w')
plt.ylabel('X[n]')
```

2)

```python
def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)

n=np.arange(0,1001,1)
h=(u(0,n)-u(9,n))
Oh=np.linspace(-np.pi,np.pi,1001)
W_Oh= np.exp(-1j)**np.dot((np.arange(h.size).reshape(h.size,1)) , (Oh.reshape(1,1001)))
H=np.dot(h, W_Oh)
absH=[abs(Xs) for Xs in H]
plt.plot(Oh,absH)

plt.title('DTFT of h[n]')
plt.xlabel('w')
plt.ylabel('H[n]')
```

3)

```python
Y1=X2*H
absY1=[abs(Ys) for Ys in Y1]
plt.plot(Oh,absY1)
plt.title('X(w)*H(w)')
plt.xlabel('w')
plt.ylabel('Y[w]')
```

4)

```python
y=signal.convolve(x2,h)
```

5)

```python
Oy=np.linspace(-np.pi,np.pi,1001)
W_Oy=np.exp(-1j)**np.dot((np.arange(y.size).reshape(y.size,1)) , (Oy.reshape(1,1001)))
Y2=np.dot(y,W_Oy)
absY2=[abs(Xs) for Xs in Y2]
plt.plot(Oy,absY2)
plt.title('Conv(x,h)')
plt.xlabel('w')
plt.ylabel('Y[w]')
```

# Part C: FIR Filter Design by Frequency Sampling

1)

```python
O=np.linspace(0,2*np.pi*(34/35),35)
HD=((O%(2*np.pi))>((2/3)*np.pi))*((O%(2*np.pi))<(2*np.pi-(2/3)*np.pi))
H=HD*np.exp(-1j*O*17)
h=np.real(fftpack.ifft(H))
plt.stem(h)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Filter Length 35')
```

2)

```python
w,H35=signal.freqz(h,1,worN=np.arange(0,2*np.pi,2*np.pi/1001))
absH35=[abs(Xs) for Xs in H35]
plt.plot(w,absH35)
plt.xlabel('w')
plt.ylabel('H[w]')
plt.title('N=35 Frequency Response')
```

3) The difference between the filter in (2) and an idea filter is that an ideal filter will have no noise around the edges of the filter, while the one in (2) does. By increasing the amount of samples, this noise decreases.

4)

```python
O=np.linspace(0,2*np.pi*(70/71),71)
HD=((O%(2*np.pi))>((2/3)*np.pi))*((O%(2*np.pi))<(2*np.pi-(2/3)*np.pi))
H=HD*np.exp(-1j*O*35)
h=np.real(fftpack.ifft(H))
plt.stem(h)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Filter Length 71')
```

```python
w,H71=signal.freqz(h,1,worN=np.arange(0,2*np.pi,2*np.pi/1001))
absH71=[abs(Xs) for Xs in H71]
plt.plot(w,absH71)
plt.xlabel('w')
plt.ylabel('H[w]')
plt.title('N=71 Frequency Response')
```

5) When the length of the filter increases the amount of noise which exists decreases. Likewise, this filter will also take 71 samples instead of 35 making it more accurate.
