```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a or sample>a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)

n=np.arange(0,10,1)

u=u(0, n)
a=np.array([1,(1/3),(-1/3)])
b=np.array([(1/9)])

h=signal.lfilter(b,a,u)

plt.stem(n,h)
plt.xlabel('n')
plt.xticks(np.arange(0,10,1))
plt.yticks([0,1/9])
plt.ylabel('h[n]')
plt.title('Unit Impulse Response')
```

# Part C: Zero State Response


```python
b=[2]
a=[1,-1,-0.1]
yn=[1,2]
n=np.arange(0,15,1)

def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)

x=(2*np.cos((2*np.pi*n)/10))*(u(0,n)-u(10,n))

y=signal.lfilter(b,a,x)

plt.stem(n,y)
plt.xlabel('n')
plt.xticks(np.arange(0,15,1))
plt.ylabel('y[n]')
plt.title('Zero State Response')
```

```python
n=np.arange(0,19,1)

def d(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a or sample>a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)

b=[2]
a=[1,-1,-0.1]

h=signal.lfilter(b,a,d(0,n))

def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)

x=(2*np.cos((2*np.pi*n)/10))*(u(0,n)-u(10,n))

y=signal.convolve(x,h)

plt.stem(np.arange(0,37,1),y)
plt.xlabel('n')
plt.xlim(0,15)
plt.ylim(-1,10)
plt.ylabel('y[n]')
plt.title('Zero State Response')
```
